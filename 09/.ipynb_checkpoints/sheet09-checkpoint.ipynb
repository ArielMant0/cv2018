## Exercise 1: Local features and interest points [4 points]

**a)** Explain in your own words: What are *local features*? How they are used?

Local features are image data that contain information regarding an image patch. They can be used to find local features for specific objects that are always present when said object is present to perform object recognition.

**b)** What are *interest points* and what are they used for? What properties are desirable?

Interest points are salient points in an image with an associated region that can be used to calculate the similarity between two images (reference image and input image) to perform tasks like object recognition.

**Desirable Properties:**

- should be *special* or *rare*, either within the image, or with respect to *common* images
- should keep positions under disruptions
- should remain in the same position with respect to the physical world in a different image of the same scene (e.g., change of viewpoint or illumination)

## Exercise 2: Computing interest points [6 points]


**a)** Explain in your own words the idea of the Moravec IP operator. What are its properties? Implement this IP operator and apply it to the image `lighthouse.png`. Try different window sizes and threshold values.

YOUR ANSWER HERE

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('images/lighthouse.png')[...,0]

window_size = 3
threshold = .2

def moravec(img, size):
    e = np.zeros(img.shape)
    c = int(size/2)
    
    for (x,y) in np.ndindex(img.shape):
        # Reference window
        ref = np.sum(img[x-c:x+c,y-c:y+c])
        
        # Perform the shifts in 4 directions (r, u, ru, rd)
        right = np.sum(img[x-c+1:x+c+1,y-c:y+c])-ref
        up = np.sum(img[x-c:x+c,y-c+1:y+c+1])-ref
        right_up = np.sum(img[x-c+1:x+c+1,y-c+1:y+c+1])-ref
        right_down = np.sum(img[x-c+1:x+c+1,y-c-1:y+c-1])-ref
        
        # Calculate the saliency
        e[x,y] = np.min([right,up,right_up,right_down])
        
    return e

plt.figure(figsize=(30,10))

saliency = moravec(img, window_size)
stacked_img = np.stack((img,)*3,axis=-1)
stacked_img[saliency > threshold] = [1,0,0]

plt.imshow(stacked_img)
```

**b)** How does the Harris corner detector work and in what sense does it improve the Moravec IP operator. Implement the Harris corner detector and apply it to `lighthouse.png`. Again, try different "window sizes" (values for $\sigma$).

YOUR ANSWER HERE

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
import cv2

k = .04
window_size = 5

img = plt.imread('images/lighthouse.png')[...,0]

def harris_corners(img, window_size, k, thresh=0):
    dy, dx = np.gradient(img)
    Ixx = dx*dx
    Ixy = dy*dx
    Iyy = dy*dy
    
    cornerList = []
    offset = int(window_size/2)
    
    for x,y in np.ndindex(img.shape):
        #Calculate sum of squares
        Sxx = np.sum(Ixx[x-offset:x+offset+1, y-offset:y+offset+1])
        Sxy = np.sum(Ixy[x-offset:x+offset+1, y-offset:y+offset+1])
        Syy = np.sum(Iyy[x-offset:x+offset+1, y-offset:y+offset+1])

        #Find determinant and trace, use to get corner response
        det = (Sxx * Syy) - (Sxy**2)
        trace = Sxx + Syy
        r = det - k*(trace**2)

        #If corner response is over threshold, add it to the corner list
        if r > thresh:
            cornerList.append((x, y))
        
    return cornerList

plt.figure(figsize=(30,10))

corners = harris_corners(img, window_size, k, 0.01)
stacked_img = np.stack((img,)*3,axis=-1)
# Color the corners
for (x,y) in corners:
    stacked_img[x,y] = [1,0,0]
    
plt.imshow(stacked_img)
```
